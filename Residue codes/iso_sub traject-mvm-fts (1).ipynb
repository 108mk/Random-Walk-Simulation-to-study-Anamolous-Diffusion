```python
#--------------------------viscoelastic fluid

#calling various library
import numpy as np
from math import *
from random import *
import matplotlib.pyplot as plt


#system parameters.
T=300.0                                    #temparature in unit kelvin
eta=1e4                           #viscosity co-efficient in unit Pa.s low frequency maxwell part
eta_s=1e-3                        #viscosity co-efficient in unit Pa.s high frequency voight part
a=2e-6                        #diameter of the probe sphere in meter


t_tot = 1e-7                        #total time in unit second
dt = t_tot/5e4                                  #time step in unit second


kb=1.38e-23                     #boltzman constant s.i unit
Gp=1000                                    #elastic modulus for both model Pa


#some constants determined from the above parameters
n=int(t_tot/dt)                               #total no of steps of the drunkard probe in 1d

D=(kb*T)/(6*pi*eta*(a/2))                  #diffusion coefficient of liquid
s=((2*D*dt)**(0.5))                        #standard deviation of the random walk in  meter
m=0.0                                      #mean of the random walk motion
lamb=(eta_s/Gp)
k=6*pi*(a/2)*Gp                            #force constant



#initial values
x1=0.0                                     #initial x cordinate in unit micro meter max
y1=0.0                                     #initial y cordinate in unit micro meter max 
x2=0.0                                     #initial x cordinate in unit micro meter voigt
y2=0.0                                     #initial y cordinate in unit micro meter voigt
t=0.0
xarray1=[]                               #list of total x after ith step
yarray1=[]                               #list of total y after ith step
xarray2=[]
yarray2=[]
dtarray=[]

#--------------------maxwell part-----------------------
i=0
for i in range(n):
    x1=x1+(gauss(m,s))
    xarray1.append(x1)
    y1=y1+(gauss(m,s))
    yarray1.append(y1)
    t=t+dt
    dtarray.append(t)
    i = i+1

    
xarraym=np.array(xarray1)
yarraym=np.array(yarray1)


j=0
for j in range(n):
    
    Ax=x2*exp(-(dt/lamb))
    Bx=sqrt(((kb*T)/k)*(1-exp(-(2*dt/lamb))))
    Ay=y2*exp(-(dt/lamb))
    By=sqrt(((kb*T)/k)*(1-exp(-(2*dt/lamb))))
    x2=(gauss(Ax,Bx))
    xarray2.append(x2)
    y2=(gauss(Ay,By))
    yarray2.append(y2)
    t=t+dt
    dtarray.append(t)
    j = j+1

xarrayv=np.array(xarray2)
yarrayv=np.array(yarray2)

```

```python
from pylab import rcParams
rcParams ['figure.figsize']= 6,6

xarray=xarraym+xarrayv
yarray=yarraym+yarrayv

plt.plot(xarray*1e9,yarray*1e9,'blue')
plt.plot(xarraym*1e9,yarraym*1e9,'red')
plt.grid()
plt.ylabel('$(nm)$',size=20)
plt.xlabel('$(nm)$',size=20)
plt.axis('square')

plt.savefig('subT_iso_1.png')
plt.show()
```
