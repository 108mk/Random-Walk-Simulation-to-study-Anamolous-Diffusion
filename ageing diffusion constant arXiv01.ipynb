```python
#---------VISCOUS FLUID-----------------SPHERICAL PARTICLE----------- VARIABLE TIME STEP-----------------
import numpy as np
from math import *
from random import *
import matplotlib.pyplot as plt

#----------PARAMETERS----------------------------------
T=300.0                                   #TEMPERATURE in unit kelvin
eta = 1e-3
a=2e-6                                    #DIAMETER of the probe sphere in meter

kb = 1.38e-23                          #boltzman constant assumption 

D_0 = (kb*T)/(6*pi*eta*a)
D_arr =[]

mean=0.0                                     #mean of the random walk motion

#---------POSITION OF PARTICLE--------------------------
x=0.0                                     #initial x cordinate in unit micro meter
y=0.0                                     #initial y cordinate in unit micro meter
x_arr=[]                                  #list of total x after ith step
y_arr=[]                                  #list of total y after ith step


vx=0.0                                     #initial x cordinate in unit micro meter
vy=0.0                                     #initial y cordinate in unit micro meter
vx_arr=[]                                  #list of total x after ith step
vy_arr=[]    
```

```python
t=0.0
t_arr=[]                               #delta time array in seconds
dt_arr=[]

age_t =0
age_t_arr =[]

tau_0 = 1e4
alpha = 1/6
```

```python
tf=1e3                             #total time in unit second
while t < tf:
    #--------------------maxwell part of the code
    p=uniform(-4,-1)
    dt=10**(p)                            #linear distributed in 10 base logscale of time
    dt_arr.append(dt)
    
    t=t+dt
    t_arr.append(t)
    
    
    age_t =  alpha*(t**(alpha-1))
    age_t_arr.append(age_t)
    
    
    D = D_0*age_t
    D_arr.append(D)
    
    
    s= 1                           #standard deviation of the random walk in  meter
    
    x = x + ((2*D)**0.5) *gauss(mean,s)*dt
    x_arr.append(x)
    
    
    y = y + (2*D)**0.5 *gauss(mean,s)*dt
    y_arr.append(y)
    
```
